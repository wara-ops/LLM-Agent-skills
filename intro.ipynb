# LLM Agent skills

In developing an agent from scratch in the previous tutorial we came across some shortcomings, some due to the naive implementation but some inherent to the strategy. In this tutorial we'll build an quite capable and easily extensible agent using _skills_ to  address some of those problems, namely:

- tools have to be written in code to extend functionality
- tools are usually 'one-trick-ponies'
- adding lots of tools consume valuable context space even if they are not used in completing a task
